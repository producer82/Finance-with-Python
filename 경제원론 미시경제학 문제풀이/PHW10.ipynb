```python
# 2. THW08 3번 문제 - 중심 극한 정리 증명

import random
import numpy as np
import matplotlib.pyplot as plt
plt.rc("font", family="Gulim")

trials = 501
samples = 1000

x = [[0 for j in range(trials)] for i in range(samples)] # 가로 trials 세로 samples -> 표본평균 구할때는 이게 편함
xBar_1 = [[0 for j in range(trials)] for i in range(samples)]
xBar_2 = [[0 for j in range(samples)] for i in range(trials)] # 가로 samples 세로 trials -> 데이터를 표시할때는 이게 편함

#난수 생성
for i in range(samples):
    for j in range(trials):
        x[i][j] = (random.random() - 1/2) * 2

#표본평균 구하기
for i in range(samples):
    for j in range(trials):
        xBar_1[i][j] = np.mean(x[i][0:j+1])

#X축 Y축 바꾸기
for i in range(samples):
    for j in range(trials):
        xBar_2[j][i] = xBar_1[i][j]

for i in range(0,501,100):
    print("n=%d"%(i))
    plt.hist(xBar_2[i], bins=100)
    plt.show()
```
